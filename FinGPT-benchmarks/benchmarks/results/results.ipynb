```python
import pandas as pd
import numpy as np
import re
```

```python
def get_name(input_string):
    # Define the regular expression pattern
    pattern = re.compile(
        r"((?:mamba|pythia)_(?:s|ms|m|l)_(?:mt\+|mt|mqsq)_\d(?:_b)?(?:_mt_\d)?)[\d_]+$"
    )

    # Search for a match in the input string
    match = pattern.search(input_string)

    # If a match is found, return the captured group
    if match:
        return match.group(1)  # Return the first (and only) capturing group
    else:
        return None
```

```python
df = pd.read_csv("wandb_export_overall2.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
# remove columns that are not needed
df = df.drop(
    columns=[
        "Created",
        "Runtime",
        "Tags",
        "Hostname",
        "State",
        "Notes",
        "User",
        "Sweep",
    ]
)
```

```python
df.head()
```

```python
# reformate the name column values to be more readable
df["Name"] = df["Name"].apply(get_name)
```

```python
df.head()
```

```python
# rename the columns old:new
df = df.rename(
    columns={
        "Name": "model",
        "fiqa_acc": "fiqa/acc",
        "fpb_acc": "fpb/acc",
        "tfns_acc": "tfns/acc",
        "nwgi_acc": "nwgi/acc",
        "arc_challenge/acc": "arc_c/acc",
        "arc_challenge/acc_stderr": "arc_c/acc_se",
        "arc_easy/acc": "arc_e/acc",
        "arc_easy/acc_stderr": "arc_e/acc_se",
        "piqa/acc": "piqa/acc",
        "winogrande/acc": "wngrd/acc",
        "winogrande/acc_stderr": "wngrd/acc_se",
        "piqa/acc_stderr": "piqa/acc_se",
        "hellaswag/acc": "hlswg/acc",
        "hellaswag/acc_stderr": "hlswg/acc_se",
        "lambada_openai/acc": "lmbd_oai/acc",
        "lambada_openai/acc_stderr": "lmbd_oai/acc_se",
        "lambada_standard/acc": "lmbd_std/acc",
        "lambada_standard/acc_stderr": "lmbd_std/acc_se",
        "lambada/acc": "lmbd/acc",
        "lambada/acc_stderr": "lmbd/acc_se",
    }
)
```

```python
df.head()
```

```python
def acc_se_format(val):
    if str(val).lower() == "n/a":
        return val
    return "±{:.2}".format(val)
    # return "±{:.2e}".format(val)

acc_se_pattern = re.compile('.*acc_se$')
acc_se_columns = [col for col in df.columns if acc_se_pattern.match(col)]
print(acc_se_columns)
df[acc_se_columns] = df[acc_se_columns].map(acc_se_format, na_action="ignore")

df.fillna("N/A", inplace=True)
df.head()
```

```python
def acc_format(val):
    if str(val).lower() == "n/a":
        return val
    return "{:.3}".format(val)

acc_pattern = re.compile('.*acc$')
acc_columns = [col for col in df.columns if acc_pattern.match(col)]
print(acc_columns)
df[acc_columns] = df[acc_columns].map(acc_format)
df.head()
```

```python
# convert column "max_len" to integer
df["max_len"] = df["max_len"].apply(lambda x: int(x) if x != "N/A" else x)
```

```python
# now split the df into two where we preserve the column 'model' and then divide the rest into two
# columns fiqa_acc	fpb_acc	tfns_acc	nwgi_acc will be in one df and the rest in another
# df = df.set_index("model")
df1 = df[["model", "max_len", "fiqa/acc", "fpb/acc", "tfns/acc", "nwgi/acc"]].sort_values("model")
df2 = df.drop(columns=["fiqa/acc", "fpb/acc", "tfns/acc", "nwgi/acc"]).sort_values("model")
```

```python
# this is great, now I want to append the acc and acc_se columns together so that we reduce the siz 
# of the dataframe so {arc_c/acc} {arc_c/acc_se} will become {arc_c/acc ± arc_c/acc_se} for each pair
# df.columns[2::2]
# df.columns[3::2]
# df_combines = df[df.columns[2::2]].astype(str) + " " + df[df.columns[3::2]].astype(str)
# df_combines

df_combines = df[["model", "max_len"]]
for c1, c2 in zip(df.columns[2::2], df.columns[3::2]):
    df_combines[c1] = df[c1].astype(str) + " ± " + df[c2].astype(str)
```

```python


# now save these as overall_fin_eval for df1 and overall_lm_eval for df2
df1.to_csv("overall_fin_eval.csv", index=False)
df2.to_csv("overall_lm_eval.csv", index=False)
```
